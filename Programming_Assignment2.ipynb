class Marsupial:                                 # Created a Class Marsupial
    def __init__(self):                          # __init__ a special function that gets called when a new object is created from a class. In this case every time a new 'Marsupial' it will give its empty pouch which is a list[]  
        self.pouch = []                          # 'self' keyword is used to access variables that belong to the object

    def put_in_pouch(self, item):                # put_in_pouch method for 'Marsupial' so that it can take an item and add it to its pouch
        self.pouch.append(item)

    def pouch_contents(self):                    # pouch_contents which allows class to show what's in the pouch 
        return self.pouch                        # sends back the list of items currently in the pouch


class Kangaroo(Marsupial):                       # Kangaroo class with 'Marsupial' as a parent class.

        self.y += dy

    def __str__(self):                           # Now, overriding a special method called __str__. This method will tell Python how to convert the object into a string for display. For our Kangaroo, we want it to say it's a Kangaroo and also tell us its current coordinates when we print it out.

```python
#....................................................... Task 5


def binary(n):
    if n == 0:                                      # If the input is zero then it will return '0', as zero in binary is 0
        return 0
    else:
        return (n % 2 + 10 * binary(int(n // 2)))   # Implemented logic for converting whole numbers to binary 

    
print(binary(0))
print(binary(1))
print(binary(1897))



```

```python
#....................................................... Task 6



from html.parser import HTMLParser                        # Html.parser is library in which, i have imported HTMLparser, It is a in-built module to analyse HTML and XML documets.

class HeadingParser(HTMLParser):                          # This class has all the functionality of HTMLparser
    def __init__(self):                                   # class's contruction method and __init__ for new objection creation
        super().__init__()                                # super() to call constructor of the superclass 'HTMLparser' to initialze the object.
        self.current_tag = None                           # Initializes an instance variable 'current_tag' to none, for processing the heading tags (h1 to h6)  
        self.indentation = 0                              # to process indentation level 'indentation' is kept zero.

    def handle_starttag(self, tag, attrs):                # It starts the reading of the html file             
        if tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:   # to check wheather the encountered tag is a heading
            self.current_tag = tag                        # if heading tag is identified then it is updated
            self.indentation = int(tag[1]) - 1            # For updation of indentation level, if the heading is 3 then it will minus one and identation level will become 2

    def handle_endtag(self, tag):                         # To encoundter ending of the html file
        if tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:   # checking the end tags, if it is a headings
            self.current_tag = None                       # current_tag set to none to indicate that we are not inside a heading

    def handle_data(self, data):                          # when heading tag is encountered
        if self.current_tag:                              # if statement for heading tag = heading
            print(' ' * self.indentation + data)          # print the heading i.e. text of that data



with open("H:\Let's Study\w3c.html", "r") as file:        # Loading theHTML file and using the parser
    html_code = file.read()                               # reading the HTML code

parser = HeadingParser()                                  # All the code which is read, now created as an object 
parser.feed(html_code)                                    # Feeding the HTML code to the parser which will analyse the material inside the file. 

```

```python
#...................................................... Task 7

import urllib.request                                                                   # Used to open and read URLs.
from bs4 import BeautifulSoup                                                           # bs4 a library in which beatifulsoup is a function to parse HTML and XML documents
from urllib.parse import urljoin                                                        # urljoin to join a base URL and another URL into an absolute URL.

def webdir(url, depth, indent):                                                         # Function webdir that takes three parameters 'url', 'dept': how many links it should follow, indent': how much to indent the output.
    print(' ' * indent + url)                                                           # Printing the current URL, preceded by spaces according to the current indentation level.
    if depth <= 1:                                                                      # If the depth is less than or equal to 1, it returns and doesn't process this URL any further.
        return
    try:                                                                                # The code inside'try' is executed and if there are any exceptions then execution jumps to the except block
        
        with urllib.request.urlopen("https://www.normalwebsite.com/") as response:      # URL is opened and gets the response from the server. The with statement ensures that the response is properly closed 
            html = response.read()                                                      # Reading the entire response from the server which is HTML
        soup = BeautifulSoup(html, 'html.parser')                                       # Finding all 'a' tags in the HTML, which is links. For each link, it runs the 'for' loop

        
        for link in soup.find_all('a'):                              
            href = link.get('href')                                                     # Getting the URL from the 'href' attribute of the 'a' tag.
            if href and not href.startswith('mailto:'):                                 # If the 'href' attribute was found and it doesn't start with 'mailto:', which is an email linkand not a webpage link.
                
                href = urljoin(url, href)                                               # Joining the base URL and the link's URL into an absolute URL
                webdir(href, depth - 1, indent + 1)                                     # Recursively calls webdir with the new URL. Decreasing depth by 1 and increasing indent by 1.
    except:                                                                             #  To catch any exceptions thrown in the try block.
        return

webdir('https://www.normalwebsite.com/', 2, 0)                                          # Calling the webdir function with a starting URL and with initial depth and indent.



```
